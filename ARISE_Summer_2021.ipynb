6) Add a column seeing if a resolution passed or failed.

```python
import numpy as np
import pandas as pd

file_path = 'sample_data/CleanedUNVotes.csv'
df1 = pd.read_csv(file_path)

df1 = df1.drop_duplicates(subset='unres', keep="first")
df1 = df1.dropna(axis=0, subset=['unres'])

df2 = df1[['unres', 'yes', 'no']]
df2

df2['Pass/Fail'] = "1"

for i in range(len(df2)):
  #average = (df2.iloc[i, 1] + df2.iloc[i, 2])/2
  if df2.iloc[i, 1] > df2.iloc[i, 2]:
    df2.iloc[i, 3] = "1"
  if df2.iloc[i, 1] < df2.iloc[i, 2]:
    df2.iloc[i, 3] = "0"

print(df2.value_counts('Pass/Fail'))
df2.to_csv('sample_data/ResDecision.csv')
```

```python
#calculate plurality

import numpy as np
import pandas as pd

file_path = 'sample_data/CleanedUNVotes.csv'
df1 = pd.read_csv(file_path)

# df1 = df1.drop_duplicates(subset='unres', keep="first")
df1 = df1.dropna(axis=0, subset=['unres'])
df2 = pd.crosstab(df1['unres'], df1['vote'])

df2['Plurality'] = -1
ctr = 0
print(df2)

for i in range(len(df2)):
  df3 = df2.iloc[[i]]
  df4 = df3.max(axis = 1)
  # print(df4)
  df4 = df4.tolist()
  # print(df4)
  
  df3_yes = df2.iloc[i, 0]
  # print(df3_yes)
  df3_no = df2.iloc[i, 2]
  df3_abstain = df2.iloc[i, 1]
  df3_notpresent = df2.iloc[i, 3]
  df3_notmember = df2.iloc[i, 4]
  
  if df4 == df3_yes:
    df2.iloc[i, 5] = 1
  if df2.iloc[i, 5] != -1 and df4 == df3_no:
    ctr = ctr + 1
  if df4 == df3_no:
    df2.iloc[i, 5] = 3
  if df2.iloc[i, 5] != -1 and df4 == df3_abstain:
    ctr = ctr + 1
  if df4 == df3_abstain:
    df2.iloc[i, 5] = 2
  if df2.iloc[i, 5] != -1 and df4 == df3_notpresent:
    ctr = ctr + 1
  if df4 == df3_notpresent:
    df2.iloc[i, 5] = 8
  if df2.iloc[i, 5] != -1 and df4 == df3_notmember:
    ctr = ctr + 1
  if df4 == df3_notmember:
    df2.iloc[i, 5] = 9

print(ctr)
print(df2.value_counts('Plurality'))

df2.to_csv('sample_data/ResDecision.csv')
print(df2.value_counts('Plurality'))
```
